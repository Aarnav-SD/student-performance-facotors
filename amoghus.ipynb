```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
import plotly as plx
import pandas as pd
import tensorflow as tf
import tensorflow.compat.v1 as tf1
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input
import ipywidgets as widgets
from IPython.display import display
from sklearn.metrics import confusion_matrix
```

```python
df = pd.read_csv('StudentPerformanceFactors.csv')
df = df.drop(['School_Type', 'Parental_Education_Level', 'Gender'], axis='columns')
df.head()
```

```python
column_list = df.columns.tolist()
column_list
```

```python
ndf = df
ndf = ndf.drop(['Hours_Studied',
 'Attendance',
 'Sleep_Hours',
 'Previous_Scores',
 'Tutoring_Sessions',
 'Physical_Activity',
 'Exam_Score',
 ], axis='columns')
ndf.head()
```

```python
columns = ndf.columns.tolist()
```

```python
for i in columns:
    ohe_df = pd.get_dummies(df[f"{i}"], prefix=f"{i}").astype(int)
    df = pd.concat([df, ohe_df], axis='columns')
    df = df.drop(i, axis='columns')
```

```python
df.head()
```

```python
X = df.drop(['Exam_Score'], axis='columns')
```

```python
model = LinearRegression()
```

```python
model.fit(X, df.Exam_Score)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, df.Exam_Score, test_size = 0.2)
```

```python
model.score(X_test, y_test)
```

## Neural Network

```python
neural_network = Sequential([
    Input(shape=(33,)),
    Dense(units = 33, activation='relu'),
    Dense(units = 18, activation='relu'),
    Dense(units = 18, activation='relu'),
    Dense(units = 11, activation='softmax')
])
neural_network.compile(optimizer='adam', loss='sparse_categorical_crossentropy')
```

```python
analysis_df = df
score_list = df['Exam_Score'].tolist()
```

```python
for i in range(len(score_list)):
    if score_list[i] == 100:
        score_list[i] = 100
    elif 100 > score_list[i] >= 90:
        score_list[i] = 90
    elif 80 <= score_list[i] < 90:
        score_list[i] = 80
    elif 70 <= score_list[i] < 80:
        score_list[i] = 70
    elif 60 <= score_list[i] < 70:
        score_list[i] = 60
    elif 50 <= score_list[i] < 60:
        score_list[i] = 50
    elif 40 <= score_list[i] < 50:
        score_list[i] = 40
    elif 30 <= score_list[i] < 40:
        score_list[i] = 30
    elif 20 <= score_list[i] < 30:
        score_list[i] = 20
    elif 10 <= score_list[i] < 20:
        score_list[i] = 10
    elif 0 <= score_list[i] < 10:
        score_list[i] = 0
analysis_df['Marks_Range'] = score_list
label_mapping = {
    0:0,
    10:1,
    20:2,
    30:3,
    40:4,
    50:5,
    60:6,
    70:7,
    80:8,
    90:9,
    100:10
}
```

```python
y_mapped = np.array([label_mapping[label] for label in analysis_df.Marks_Range])
```

```python
neural_network.fit(X, y_mapped, epochs=300)
```

```python
predictions = neural_network.predict(X)
```

```python
predicted_class = np.argmax(predictions, axis=1)
predicted_class
```

```python
marks_range=[]
for i in predicted_class:
    for key, value in label_mapping.items():
        if i == value:
            marks_range.append(key)
```
